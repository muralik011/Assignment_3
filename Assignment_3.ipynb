<b>5. How many global scopes are there in a Python program? How many local scopes?

There is only one global scope in python program. Each function has it's own local scope. Variables in global scope can be called inside a function provided there is no variable with the same name in the local scope of a function. A variable declared inside a function is not available outside the function.
